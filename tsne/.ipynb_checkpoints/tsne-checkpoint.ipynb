```python
def label_mask(labels):
    mask = (labels < 0) + (labels > 18)
    labels[mask] = 19

    return labels
```

```python
labels = label_mask(labels)
```

```python
x = np.arange(129)
y = np.arange(65)

X, Y = np.meshgrid(x,y)
plt.pcolor(X,Y, labels.reshape(65, 129), vmin = 0, vmax = 20)
```
